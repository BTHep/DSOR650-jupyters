```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from PIL import Image

#/Users/benhepner/Downloads/BirdPolaris.jpg
```

```python

def apply_pca_and_visualize(image_path, components_list):
    # Load the image
    img = Image.open(image_path)
    
    # Convert image to grayscale
    gray_img = img.convert('L')
    
    # Convert grayscale image to numpy array
    img_array = np.array(gray_img)
    
    # Display the original grayscale image
    plt.figure(figsize=(6, 6))
    plt.imshow(img_array, cmap='gray')
    plt.title('Original Grayscale Image')
    plt.axis('off')  # Hide axis
    plt.show()
    
    # Normalize the image array for PCA
    img_normalized = img_array / 255.0
    
    for n_components in components_list:
        pca = PCA(n_components=n_components)
        
        # Apply PCA on each row
        pca.fit(img_normalized)
        img_transformed = pca.transform(img_normalized)
        img_reconstructed = pca.inverse_transform(img_transformed) * 255
        
        # Visualize the reconstructed image
        plt.figure(figsize=(6, 6))
        plt.imshow(img_reconstructed, cmap='gray', vmin=0, vmax=255)
        plt.title(f'{n_components} Principal Components')
        plt.axis('off')  # Hide axis
        plt.show()


# Example usage:
apply_pca_and_visualize('/Users/benhepner/Downloads/BirdPolaris.jpg', [1, 10, 50, 100, 500])
```

```python
image_path = '/Users/benhepner/Downloads/BirdPolaris.jpg'
img = Image.open(image_path)
gray_img = img.convert('L')  # Convert to grayscale
img_array = np.array(gray_img)
img_normalized = img_array / 255.0  # Normalize the image data

# Fit PCA
pca = PCA()
pca.fit(img_normalized)

# Check the number of components
num_components = pca.n_components_
print(f"Number of principal components: {num_components}")
```
