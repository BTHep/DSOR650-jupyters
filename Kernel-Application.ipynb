```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def apply_kernel_and_export(image_path, kernel, export_path):
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Image cannot be loaded. Please check the path.")
        return
    
    # Convert image to RGB (OpenCV uses BGR by default)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Apply the kernel to the image
    filtered_image = cv2.filter2D(image_rgb, -1, kernel)
    
    # Display the original and filtered images
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(filtered_image)
    plt.title('Filtered Image')
    plt.axis('off')
    
    plt.show()
    
    # Export the filtered image
    # Convert filtered_image back to BGR before saving
    filtered_image_bgr = cv2.cvtColor(filtered_image, cv2.COLOR_RGB2BGR)
    cv2.imwrite(export_path, filtered_image_bgr)
    print(f"Filtered image exported to {export_path}")

# Example usage
if __name__ == "__main__":
    image_path = '/Users/benhepner/Downloads/AFIT.jpeg' # Update this with the path to your image
    export_path = '/Users/benhepner/Downloads/AFIT_Filtered.jpeg' # Path where you want to save the filtered image
    # Define a 3x3 kernel. This is an example kernel for edge detection.
    kernel = np.array([[3, 1/9, 0],
                       [0,  100, 0],
                       [0, 1/9, 1/9]])
    
    apply_kernel_and_export(image_path, kernel, export_path)
```
